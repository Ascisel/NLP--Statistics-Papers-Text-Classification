#### Distribution of classes for binary classification

```python
terms
```

```python
n_examples = counts.sum()
```

```python
classes = ['stat.ML', 'stat.TH', 'stat.CO', 'stat.ME', 'stat.OT']
classes_count = {}

for cl in classes:
    index = [cl in term for term in terms]
    classes_count[cl] = counts[index].sum()
    print(f'Balance ratio for {cl}: {classes_count[cl]/n_examples}')
```

```python
weights = {}
for cl in classes:
    weights_cl = {}
    w_1 = n_examples/(2*classes_count[cl])
    w_0 = n_examples/(2*(n_examples -classes_count[cl]))
    weights_cl[1] = w_1
    weights_cl[0] = w_0
    weights[cl] = weights_cl

weights
```

We see that we need to deal with unbalanced dataset. Biggest problem is with 'stat.OT' class where we got very few examples. Need to think about dealing with unbalance such as augmentation, resampling, using class weight.

#### Unique targets in our dataset

```python
terms.shape[0]
```

Max number of possible classes in our problem when try to use all permutations:

$1+5+\binom{5}{2}+\binom{5}{3}+\binom{5}{4}+1=5+1+10+10+5+1=32$

So if we use multiclass classifier for every possible permutation we need to have 32 possible target classes, but we only got 23 in our training data. Our dataset seems to be very small if we want to use multiclass, we would have so little examples of each class that overfitting could be problematic.
